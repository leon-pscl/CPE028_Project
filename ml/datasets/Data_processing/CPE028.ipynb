```python
import random
import pandas as pd

random.seed(42)

devices = ["phone", "tablet", "laptop", "smartwatch"]

battery_actions = [
    "drains quickly",
    "loses charge overnight",
    "drops from full charge within hours",
    "dies unexpectedly",
    "does not hold a charge",
    "gets hot while charging",
    "discharges while idle",
    "loses battery health rapidly",
    "powers off suddenly",
    "takes too long to charge"
]

battery_contexts = [
    "while gaming",
    "during video calls",
    "after a software update",
    "when using GPS",
    "during normal use",
    "while watching videos",
    "when connected to WiFi",
    "during standby",
    "while charging",
    "during online meetings"
]



hardware_parts = [
    "keyboard",
    "speaker",
    "microphone",
    "USB port",
    "charging port",
    "camera",
    "trackpad",
    "fingerprint sensor",
    "power button",
    "motherboard"
]

hardware_failures = [
    "stopped working",
    "works intermittently",
    "is not detected",
    "produces unusual behavior",
    "fails randomly",
    "became unresponsive",
    "is physically damaged",
    "disconnects frequently",
    "shows signs of failure",
    "malfunctions during use"
]


software_problems = [
    "apps keep crashing",
    "system freezes randomly",
    "device restarts unexpectedly",
    "operating system fails to boot",
    "software update failed",
    "WiFi disconnects frequently",
    "Bluetooth stops working",
    "device enters a boot loop",
    "screen freezes",
    "settings reset automatically"
]

software_contexts = [
    "after updating",
    "during startup",
    "while using multiple apps",
    "when opening settings",
    "during normal operation",
    "after installing new software",
    "when connected to the internet",
    "during file transfers",
    "after rebooting",
    "while charging"
]


screen_issues = [
    "screen flickers",
    "display shows vertical lines",
    "touchscreen is unresponsive",
    "brightness changes automatically",
    "screen goes black",
    "ghost touches occur",
    "screen freezes",
    "display shows distorted colors",
    "dead pixels appeared",
    "screen cracked"
]

screen_contexts = [
    "after a drop",
    "during normal use",
    "while charging",
    "after replacing the screen",
    "during gaming",
    "when using social media",
    "after an update",
    "during video playback",
    "when exposed to sunlight",
    "randomly throughout the day"
]


damage_events = [
    "fell into water",
    "was dropped on concrete",
    "was exposed to heavy rain",
    "had coffee spilled on it",
    "fell from a table",
    "was crushed inside a bag",
    "was left inside a hot car",
    "was dropped down stairs",
    "was exposed to beach sand",
    "was hit during a vehicle accident"
]

damage_results = [
    "and no longer turns on",
    "and the screen shattered",
    "and the touchscreen stopped working",
    "and charging no longer works",
    "and the speaker became distorted",
    "and random shutdowns started occurring",
    "and the display became damaged",
    "and internal components failed",
    "and the battery started overheating",
    "and several buttons stopped responding"
]


def generate_battery(n=400):
    items = []
    while len(items) < n:
        items.append(
            f"My {random.choice(devices)} {random.choice(battery_actions)} "
            f"{random.choice(battery_contexts)}."
        )
    return items


def generate_hardware(n=400):
    items = []
    while len(items) < n:
        items.append(
            f"The {random.choice(hardware_parts)} on my "
            f"{random.choice(devices)} {random.choice(hardware_failures)}."
        )
    return items


def generate_software(n=400):
    items = []
    while len(items) < n:
        items.append(
            f"My {random.choice(devices)} "
            f"{random.choice(software_problems)} "
            f"{random.choice(software_contexts)}."
        )
    return items


def generate_screen(n=400):
    items = []
    while len(items) < n:
        items.append(
            f"The {random.choice(screen_issues)} "
            f"{random.choice(screen_contexts)}."
        )
    return items


def generate_damage(n=400):
    items = []
    while len(items) < n:
        items.append(
            f"My {random.choice(devices)} "
            f"{random.choice(damage_events)} "
            f"{random.choice(damage_results)}."
        )
    return items


dataset = []

for x in generate_battery():
    dataset.append([x, "Battery degradation"])

for x in generate_hardware():
    dataset.append([x, "Hardware issue"])

for x in generate_software():
    dataset.append([x, "Software issue"])

for x in generate_screen():
    dataset.append([x, "Cracked screen"])

for x in generate_damage():
    dataset.append([x, "Water/Liquid damage"])

random.shuffle(dataset)

df = pd.DataFrame(dataset, columns=["issue", "label"])

print("Total rows:", len(df))
print(df.sample(10))

df.to_csv("device_issue_dataset_2000.csv", index=False)
```

```python
import re
import nltk
from sklearn.feature_extraction.text import CountVectorizer

nltk.download('punkt')

def categorize_issue_text(issue_description):
    issue_description = issue_description.lower()

    if re.search(r'battery|charge|drains|loses charge|dies|hot while charging|discharges', issue_description):
        return "Battery degradation"
    if re.search(r'keyboard|speaker|microphone|usb port|charging port|camera|trackpad|fingerprint sensor|power button|motherboard|stopped working|intermittently|not detected|malfunctions|hardware', issue_description):
        return "Hardware issue"
    if re.search(r'apps crashing|system freezes|restarts unexpectedly|fails to boot|update failed|wifi disconnects|bluetooth stops|boot loop|screen freezes|settings reset|software|os|operating system', issue_description):
        return "Software issue"
    if re.search(r'screen|display|flickers|lines|unresponsive|black screen|ghost touches|distorted colors|dead pixels|cracked', issue_description):
        return "Cracked screen"
    if re.search(r'water|spilled|liquid|dropped on concrete|exposed to rain|fell from table|crushed|hot car|dropped down stairs|beach sand|hit during accident', issue_description):
        return "Water/Liquid damage"
    return "Unknown"

categorized_to_label = to_label.apply(categorize_issue_text)

print("Original 'to_label' (sample):")
print(to_label.head())
print("\nCategorized 'to_label' (sample):")
print(categorized_to_label.head())
print("\nValue counts of categorized 'to_label':")
print(categorized_to_label.value_counts())
```

```python
data['predicted_label'] = categorized_to_label

print("DataFrame with 'predicted_label' column:")
display(data.head())
```

```python
data = data[['issue', 'predicted_label']]
print("DataFrame after dropping 'label' column:")
display(data.head())
```
